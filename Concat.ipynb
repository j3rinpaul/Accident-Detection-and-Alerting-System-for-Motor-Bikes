```python
import numpy as np
import pandas as pd
```

```python
acc_Normal = pd.read_csv("NORMAL\Accelerometer.csv")
```

```python
gyro_Normal = pd.read_csv("NORMAL\Gyroscope.csv")
```

```python
acc_Rash = pd.read_csv("RASH\Accelerometer.csv")
```

```python
gyro_Rash = pd.read_csv("RASH\Gyroscope.csv")
```

```python
acc_Normal
```

```python
x_acc_Normal = np.array([ abs(acc_Normal['X'][i] - acc_Normal['X'][i+1]) for i in range(len(acc_Normal['X'])-1) ])
x_acc_Rash = np.array([ abs(acc_Rash['X'][i] - acc_Rash['X'][i+1]) for i in range(len(acc_Rash['X'])-1) ])
x_gyro_Normal = np.array([abs(gyro_Normal['X'][i]) for i in range(len(gyro_Normal['X']))])
x_gyro_Rash = np.array([abs(gyro_Rash['X'][i]) for i in range(len(gyro_Rash['X']))])
y_acc_Normal = np.array([ abs(acc_Normal['Y'][i] - acc_Normal['Y'][i+1]) for i in range(len(acc_Normal['Y'])-1) ])
y_acc_Rash = np.array([ abs(acc_Rash['Y'][i] - acc_Rash['Y'][i+1]) for i in range(len(acc_Rash['Y'])-1) ])
y_gyro_Normal = np.array([abs(gyro_Normal['Y'][i]) for i in range(len(gyro_Normal['Y']))])
y_gyro_Rash = np.array([abs(gyro_Rash['Y'][i]) for i in range(len(gyro_Rash['Y']))])
z_acc_Normal = np.array([ abs(acc_Normal['Z'][i] - acc_Normal['Z'][i+1]) for i in range(len(acc_Normal['Z'])-1) ])
z_acc_Rash = np.array([ abs(acc_Rash['Z'][i] - acc_Rash['Z'][i+1]) for i in range(len(acc_Rash['Z'])-1) ])
z_gyro_Normal = np.array([abs(gyro_Normal['Z'][i]) for i in range(len(gyro_Normal['Z']))])
z_gyro_Rash = np.array([abs(gyro_Rash['Z'][i]) for i in range(len(gyro_Rash['Z']))])


x_acc_Accident = 2*x_acc_Rash
x_gyro_Accident = 2*x_gyro_Rash
y_acc_Accident = 2*y_acc_Rash
y_gyro_Accident = 2*y_gyro_Rash
z_acc_Accident = 2*z_acc_Rash
z_gyro_Accident = 2*z_gyro_Rash

x_acc_Rash.shape                                 
```

```python
x_acc_Normal = np.mean(x_acc_Normal[0:3045].reshape(-1,5),axis=1)
x_acc_Rash = np.mean(x_acc_Rash[0:1335].reshape(-1,5),axis=1)
x_acc_Accident = np.mean(x_acc_Accident[0:1335].reshape(-1,5),axis=1)
x_gyro_Normal = np.mean(x_gyro_Normal[0:3045].reshape(-1,5),axis=1)
x_gyro_Rash = np.mean(x_gyro_Rash[0:1335].reshape(-1,5),axis=1)
x_gyro_Accident = np.mean(x_gyro_Accident[0:1335].reshape(-1,5),axis=1)
y_acc_Normal = np.mean(y_acc_Normal[0:3045].reshape(-1,5),axis=1)
y_acc_Rash = np.mean(y_acc_Rash[0:1335].reshape(-1,5),axis=1)
y_acc_Accident = np.mean(y_acc_Accident[0:1335].reshape(-1,5),axis=1)
y_gyro_Normal = np.mean(y_gyro_Normal[0:3045].reshape(-1,5),axis=1)
y_gyro_Rash = np.mean(y_gyro_Rash[0:1335].reshape(-1,5),axis=1)
y_gyro_Accident = np.mean(y_gyro_Accident[0:1335].reshape(-1,5),axis=1)
z_acc_Normal = np.mean(z_acc_Normal[0:3045].reshape(-1,5),axis=1)
z_acc_Rash = np.mean(z_acc_Rash[0:1335].reshape(-1,5),axis=1)
z_acc_Accident = np.mean(z_acc_Accident[0:1335].reshape(-1,5),axis=1)
z_gyro_Normal = np.mean(z_gyro_Normal[0:3045].reshape(-1,5),axis=1)
z_gyro_Rash = np.mean(z_gyro_Rash[0:1335].reshape(-1,5),axis=1)
z_gyro_Accident = np.mean(z_gyro_Accident[0:1335].reshape(-1,5),axis=1)

z_gyro_Rash.shape
```

```python
for i in range(2):
    x_acc_Normal = np.concatenate((x_acc_Normal,x_acc_Normal),axis = 0)
    x_gyro_Normal = np.concatenate((x_gyro_Normal,x_gyro_Normal),axis = 0)
    y_acc_Normal = np.concatenate((y_acc_Normal,y_acc_Normal),axis = 0)
    y_gyro_Normal = np.concatenate((y_gyro_Normal,y_gyro_Normal),axis = 0)
    z_acc_Normal = np.concatenate((z_acc_Normal,z_acc_Normal),axis = 0)
    z_gyro_Normal = np.concatenate((z_gyro_Normal,z_gyro_Normal),axis = 0)
for i in range(3):
    x_acc_Rash = np.concatenate((x_acc_Rash,x_acc_Rash),axis = 0)
    x_gyro_Rash = np.concatenate((x_gyro_Rash,x_gyro_Rash),axis = 0)
    y_acc_Rash = np.concatenate((y_acc_Rash,y_acc_Rash),axis = 0)
    y_gyro_Rash = np.concatenate((y_gyro_Rash,y_gyro_Rash),axis = 0)
    z_acc_Rash = np.concatenate((z_acc_Rash,z_acc_Rash),axis = 0)
    z_gyro_Rash = np.concatenate((z_gyro_Rash,z_gyro_Rash),axis = 0)
for i in range(3):
    x_acc_Accident = np.concatenate((x_acc_Accident,x_acc_Accident),axis = 0)
    x_gyro_Accident = np.concatenate((x_gyro_Accident,x_gyro_Accident),axis = 0)
    y_acc_Accident = np.concatenate((y_acc_Accident,y_acc_Accident),axis = 0)
    y_gyro_Accident = np.concatenate((y_gyro_Accident,y_gyro_Accident),axis = 0)
    z_acc_Accident = np.concatenate((z_acc_Accident,z_acc_Accident),axis = 0)
    z_gyro_Accident = np.concatenate((z_gyro_Accident,z_gyro_Accident),axis = 0)

x_acc_Normal = x_acc_Normal[0:2000]
x_gyro_Normal = x_gyro_Normal[0:2000]
y_acc_Normal = y_acc_Normal[0:2000]
y_gyro_Normal = y_gyro_Normal[0:2000]
z_acc_Normal = z_acc_Normal[0:2000]
z_gyro_Normal = z_gyro_Normal[0:2000]

x_acc_Rash = x_acc_Rash[0:2000]
x_gyro_Rash = x_gyro_Rash[0:2000]
y_acc_Rash = y_acc_Rash[0:2000]
y_gyro_Rash = y_gyro_Rash[0:2000]
z_acc_Rash = z_acc_Rash[0:2000]
z_gyro_Rash = z_gyro_Rash[0:2000]

x_acc_Accident = x_acc_Accident[0:2000]
x_gyro_Accident = x_gyro_Accident[0:2000]
y_acc_Accident = y_acc_Accident[0:2000]
y_gyro_Accident = y_gyro_Accident[0:2000]
z_acc_Accident = z_acc_Accident[0:2000]
z_gyro_Accident = z_gyro_Accident[0:2000]
```

```python
DATA = {'X_acc' : np.round_(np.concatenate((x_acc_Normal,x_acc_Rash,x_acc_Accident),axis=0),decimals = 2),
        'Y_acc' : np.round_(np.concatenate((y_acc_Normal,y_acc_Rash,y_acc_Accident),axis=0),decimals = 2),
        'Z_acc' : np.round_(np.concatenate((z_acc_Normal,z_acc_Rash,z_acc_Accident),axis=0),decimals = 2),
        'X_gyro': np.round_(np.concatenate((x_gyro_Normal,x_gyro_Rash,x_gyro_Accident),axis=0),decimals = 2),
        'Y_gyro': np.round_(np.concatenate((y_gyro_Normal,y_gyro_Rash,y_gyro_Accident),axis=0),decimals = 2),
        'Z_gyro': np.round_(np.concatenate((z_gyro_Normal,z_gyro_Rash,z_gyro_Accident),axis=0),decimals = 2),
        'Label' : np.concatenate((np.array(['Normal' for i in range(2000)]),np.array(['Rash' for i in range(2000)]),np.array(['Accident' for i in range(2000)])))
       }
```

```python
Data = pd.DataFrame(DATA)
Data
```

```python
Data.to_csv('ALL_DATA4.csv',index = False)
```
